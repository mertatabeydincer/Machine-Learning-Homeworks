```
import numpy as np
import pandas as pd
from random import randrange
 
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

columnNames = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Output"]
df= pd.read_csv(url) #Reads dataset from url
rows = len(df)
columns = len(df.columns)
```

```
df.columns = columnNames #Assings the column names
print(df.to_string()) #Prints all dataset
```

```
print(df.shape) #Prints the shape (dimensions) of data set in (row,column) format
```

```
for col in df.columns: #Prints all column names
    print(col)
```
