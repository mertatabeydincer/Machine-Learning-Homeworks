```
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
```

```
url="https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"

columnNames = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth", "Output"]
df = pd.read_csv(url, index_col=None, header=None, names=columnNames) #Reads dataset from url
```

```
attributeCols = ["SepalLength", "SepalWidth", "PetalLength", "PetalWidth"]

X = df[attributeCols].to_numpy() #Seperate features from labels.
#We will be using only inputs since this is a unsupervised learning method.
```

```
from warnings import filterwarnings
filterwarnings('ignore')
numberOfClusters = [1,3,10,50,149]

for i in numberOfClusters:
  print("\n\nNUMBER OF CLUSTERS (n_clusters): {}".format(i))
  for j in range (1,11):
    kmeans = KMeans(n_clusters=i, n_init=j, random_state=5).fit(X)
    #print("\nNumber of runs (n_init): {}".format(j))                        #These prints have been used to figure out if n_init has any effect on output.
    #print("\nLABELS:\n{}".format(kmeans.labels_[:])                        #These helps to idenify eack n_init value's effect on results.
    #print("\nCLUSTER CENTERS:\n{}".format(kmeans.cluster_centers_[:]))
    #print("\nINERTIA: {:.3f}".format(kmeans.inertia_))
    #print("\n-------------------------------------------------------------------")
 
  print("\nLABELS:\n{}".format(kmeans.labels_[:]))   
  print("\nCLUSTER CENTERS:\n{}".format(kmeans.cluster_centers_[:]))
  print("\nINERTIA: {:.3f}".format(kmeans.inertia_))

  print("\n-------------------------------------------------------------------")
```
